### Proses

```python
AttFla = list(df_baru["Attrition_Flag"].drop_duplicates())# ["Attrited Customer", "Existing Customer"]
GenCus = list(df_baru["Gender"].drop_duplicates())# ["M", "F"]
EduLev = list(df_baru["Education_Level"].drop_duplicates())# ["High School","Graduate","Uneducated","Post-Graduate","Doctorate","Unknown","College"]
MarSta = list(df_baru["Marital_Status"].drop_duplicates())# ["Single","Married","Unknown"]
IncCat = list(df_baru["Income_Category"].drop_duplicates())# ['$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Less than $40K', 'Unknown']
CarCat = list(df_baru["Card_Category"].drop_duplicates())# ['Blue']

Att_Customer = len(df_baru[df_baru["Attrition_Flag"] == "Attrited Customer" ])#10
Exi_Customer = len(df_baru[df_baru["Attrition_Flag"] == "Existing Customer" ])#10
```

```python
def find_colm(column):
  kolom = list(df_baru.columns) # isinya nama kolomnya ['Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']
  colm = [] # ini kolom yang ada di variabel x 
  
  #proses mencari x ini termasuk dalam kolom apa
  for i in column:
    if i in AttFla:
      if "Attrition_Flag" not in colm:
        colm.append("Attrition_Flag")    
    elif i in GenCus: 
      if "Gender" not in colm:
        colm.append("Gender")
    elif i in EduLev or i in IncCat: #alasan menggunakan or , karena ada memiliki nilai yg sama yaitu "Unknown"
      if "Education_Level" not in colm:
        colm.append("Education_Level")
      elif "Income_Category" not in colm and i in IncCat:
        colm.append("Income_Category")
    elif i in MarSta:
      if "Marital_Status" not in colm:
        colm.append("Marital_Status")
    elif i in CarCat:
      if "Card_Category" not in colm:
        colm.append("Card_Category")
    else:
      break

  return colm
  """
  # zz melihat kolom apa yg nda ada
  # jelas sudah Attrition_Flag tidak ada, jadi tidak ditampilkan
   zz = "" 
   for i in kolom:
     if i not in colm:
       zz = i
  print(zz) # Attrition_Flag
  """
```

```python
def count_row(colm):# function untuk mencari jumlah baris sesuai kondisi
  att_lis=[] # [3, 2, 6, 1, 10] list berisi jumlah dari Attrited Customer
  exi_lis=[] # [6, 3, 5, 3, 10] Existing Customer
  idxn = 0
  while idxn < len(x): 
    adx = df_baru[(df_baru["Attrition_Flag"]== "Attrited Customer") & (df_baru[colm[idxn]]== x[idxn]) ]
    ady = df_baru[(df_baru["Attrition_Flag"]== "Existing Customer") & (df_baru[colm[idxn]]== x[idxn]) ]
    pan = len(adx)
    pam = len(ady)
    att_lis.append(pan)
    exi_lis.append(pam)
    idxn += 1

  all_lis=[att_lis,exi_lis]
  
  return all_lis
  # print(att_lis)
  # print(exi_lis)
  # print(all_lis)
```

```python
def count_value(lis): # setelah dari func count_row() jumlah baris tersebut kemudian dibagi banyak baris Attrited Customer dan Existing Customer

  att2_lis=[] # hasil dari perhitungan dibawah [0.3, 0.2, 0.6, 0.1, 1.0]
  exi2_lis=[] #
  for at in lis[0]:#[3, 2, 6, 1, 10]
    att_val = at / Att_Customer # x / 10
    att2_lis.append(att_val)
  
  for ex in lis[1]:#[6, 3, 5, 3, 10]
    exi_val = ex / Exi_Customer
    exi2_lis.append(exi_val)

  all2_lis=[att2_lis,exi2_lis]
  
  return all2_lis
  # print(att2_lis)
  # print(exi2_lis)
```

```python
def final_value(lis):
  att_val = [] #0.0036 dapat dari perulangan for dibawah
  exi_val = []

  #dibawah (bagian perulangan for) merupakan proses perkalian dari hasil count_value() [0.3, 0.2, 0.6, 0.1, 1.0] 
  for y in lis[0]:
      if not att_val:
          att_val.append(y)
      else:
          att_val[0] *= y
  for a in lis[1] :
      if not exi_val:
          exi_val.append(a)
      else:
          exi_val[0] *= a

  # menghitung P(X|Ci) * P (Ci)
  att_final = att_val[0] * Att_Customer / len(df_baru) # x * 10 / 20
  exi_final = exi_val[0] * Exi_Customer / len(df_baru)


  if att_final > exi_final:
    # print(att_final)
    value_final = ("Attrited Customer", att_final)

  else:
    # print(exi_final)
    value_final = ("Existing Customer", exi_final)
  
  print(value_final)
```

```python
def naive_bayes(list_naive):
  cari_kolom = find_colm(list_naive)
  hitung_baris = count_row(cari_kolom)
  hitung_nilai = count_value(hitung_baris)
  final_value(hitung_nilai)
```
